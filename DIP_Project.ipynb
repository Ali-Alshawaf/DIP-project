# Challenge 1

```python
import cv2
import matplotlib.pyplot as plt

# Load the image
img = cv2.imread('images/chall_1.jpg')

# Display the original image
fig, axs = plt.subplots(1, 2, figsize=(20, 5))
axs[0].imshow(img)
axs[0].set_title('Cracked street')
axs[0].axis('off')

# Convert the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply Gaussian blur to reduce noise
blur = cv2.GaussianBlur(gray, (5, 5), 0)

# Apply Canny edge detection
edges = cv2.Canny(blur, 30, 150)

# Find contours
contours, hierarchy = cv2.findContours(edges, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# Draw contours on the original image
cv2.drawContours(img, contours, -1, (0, 255, 0), 2)

# Display the result image 
axs[1].imshow(img)
axs[1].set_title('Cracked street areas')
axs[1].axis('off')
plt.show
```

# Challenge 2

```python
import cv2
import matplotlib.pyplot as plt

# Load the image
img = cv2.imread("images/chall_2.jpg")

# Display the original image
fig, axs = plt.subplots(1, 2, figsize=(20, 5))
axs[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axs[0].set_title('Park')
axs[0].axis('off')

# Convert the image to grayscale
grayscale= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply a blur to the image to smooth out the edges
blur = cv2.blur(grayscale, (5, 5))

# Apply a threshold to the image to convert it to a binary image
threshold = cv2.threshold(blur, 150, 255, cv2.THRESH_BINARY)[1]

# Find the contours in the binary image
conts, hierarchy = cv2.findContours(threshold, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# Draw the contours on the original image
cv2.drawContours(img, conts, -1, (30,144,255), 2)

# Display the result image with the detected contours
axs[1].imshow(img)
axs[1].set_title('Distinct area of park')
axs[1].axis('off')
plt.show
```

# Challenge 3

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('images/chall_3.jpg')
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

lower_green = np.array([25, 52, 72])
upper_green = np.array([102, 255, 255])

mask = cv2.inRange(hsv, lower_green, upper_green)
res = cv2.bitwise_and(img,img, mask= mask)

fig, axs = plt.subplots(1, 2, figsize=(20, 5))
axs[0].imshow(img)
axs[0].set_title('sick plant')
axs[0].axis('off')
# Display the result image 
axs[1].imshow(res)
axs[1].set_title('disease areas of plant')
axs[1].axis('off')
plt.show
```
